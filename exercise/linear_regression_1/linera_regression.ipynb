# Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn import linear_model

```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
df
```

```python
df.head()
```

```python
df
```

```python
%matplotlib inline
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('$','')
df.columns
#df.rename(columns={percapitaincome(US$):capita},inplace=True)
plt.xlabel("year")
plt.ylabel("income")
plt.scatter(df.year,df.per_capita_income_us)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.per_capita_income_us)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
n=reg.predict([[2020]])

df.append({'year': 2020,'per_capita_income_us':n[0]}, ignore_index=True)
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.per_capita_income_us,color="red",marker="+")
plt.plot(df.year,reg.predict(df[['year']]),color="blue")

```

```python
import pickle
with open("model_pickle","wb") as f:
   pickle.dump(reg,f)
```
